```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import glob
import pandas as pd
import numpy as np

df = pd.concat([pd.read_csv(f, sep=',', na_values=".", header=None) for f in glob.glob('watch/gyro/*.txt')], ignore_index=True)

Bfx = df[[3]][df[1] == 'B']
Bfy = df[[4]][df[1] == 'B']
Bfz = df[[5]][df[1] == 'B']

Ffx = df[[3]][df[1] == 'F']
Ffy = df[[4]][df[1] == 'F']
Ffz = df[[5]][df[1] == 'F']

Sfx = df[[3]][df[1] == 'S']
Sfy = df[[4]][df[1] == 'S']
Sfz = df[[5]][df[1] == 'S']

fig = plt.figure()
#######################################################################
ax = fig.add_subplot(231, projection='3d')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

x1 = Bfx
y1 = Bfy
z1 = Bfz

ax.scatter(x1, y1, z1, c='g', marker='x')

x2 = Ffx
y2 = Ffy
z2 = Ffz

ax.scatter(x2, y2, z2, c='b', marker='v')

x3 = Sfx
y3 = Sfy
z3 = Sfz

ax.scatter(x3, y3, z3, c='r', marker='o')

######################################################################

ax2 = fig.add_subplot(232, projection='3d')

ax2.scatter(x1, y1, z1, c='g', marker='x')

ax2.set_xlabel('X Label')
ax2.set_ylabel('Y Label')
ax2.set_zlabel('Z Label')

ax2.scatter(x2, y2, z2, c='b', marker='v')

ax2.scatter(x3, y3, z3, c='r', marker='o')

ax2.view_init(90, 225)

######################################################################

ax3 = fig.add_subplot(233, projection='3d')

ax3.scatter(x1, y1, z1, c='g', marker='x')

ax3.set_xlabel('X Label')
ax3.set_ylabel('Y Label')
ax3.set_zlabel('Z Label')

ax3.scatter(x2, y2, z2, c='b', marker='v')

ax3.scatter(x3, y3, z3, c='r', marker='o')

ax3.view_init(180, 270)

#######################################################################

ax4 = fig.add_subplot(234, projection='3d')

ax4.scatter(x1, y1, z1, c='g', marker='x')

ax4.set_xlabel('X Label')
ax4.set_ylabel('Y Label')
ax4.set_zlabel('Z Label')

ax4.scatter(x2, y2, z2, c='b', marker='v')

ax4.scatter(x3, y3, z3, c='r', marker='o')

ax4.view_init(180, 45)

########################################################################

ax5 = fig.add_subplot(235, projection='3d')

ax5.scatter(x1, y1, z1, c='g', marker='x')

ax5.set_xlabel('X Label')
ax5.set_ylabel('Y Label')
ax5.set_zlabel('Z Label')

ax5.scatter(x2, y2, z2, c='b', marker='v')

ax5.scatter(x3, y3, z3, c='r', marker='o')

ax5.view_init(270, 90)

#######################################################################

ax6 = fig.add_subplot(236, projection='3d')

ax6.scatter(x1, y1, z1, c='g', marker='x')

ax6.set_xlabel('X Label')
ax6.set_ylabel('Y Label')
ax6.set_zlabel('Z Label')

ax6.scatter(x2, y2, z2, c='b', marker='v')

ax6.scatter(x3, y3, z3, c='r', marker='o')

ax6.view_init(315, 315)

########################################################################

plt.show()
```
